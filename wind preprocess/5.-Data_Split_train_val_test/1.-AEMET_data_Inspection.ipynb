#### Multiple plot

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 4))

time_index_p = 501

ws_single_data_k = ws_data_k[time_index_p, :, :]
ws_single_data_m = ws_data_k[time_index_p, :, :]

pc1 = ax1.pcolormesh(lon_k, lat_k, ws_single_data_k, cmap='gist_heat', vmin = 0, vmax = 50)
plt.colorbar(pc1, ax=ax1, label='Wind Speed (km/h)')
ax1.set_xlabel('lon')
ax1.set_ylabel('lat')
ax1.set_title('Pcolormesh')

pc2 = ax2.pcolormesh(lon_k, lat_k, ws_single_data_m, cmap='gist_heat', vmin = 0, vmax = 50)
plt.colorbar(pc2, ax=ax2, label='Wind Speed (km/s)')
ax2.set_xlabel('lon')
ax2.set_ylabel('lat')
ax2.set_title('Pcolormesh')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 4))

time_index_c = 515

ws_single_data_k = ws_data_k[time_index_c, :, :]
ws_single_data_m = ws_data_k[time_index_c, :, :]

pc1 = ax1.contourf(lon_k, lat_k, ws_single_data_k, 10, cmap='gist_heat', vmin = 0, vmax = 50)
plt.colorbar(pc1, ax=ax1, label='Wind Speed (km/h)')
ax1.set_xlabel('lon')
ax1.set_ylabel('lat')
ax1.set_title('Contourf')

pc2 = ax2.contourf(lon_k, lat_k, ws_single_data_m, 10, cmap='gist_heat', vmin = 0, vmax = 50)
plt.colorbar(pc2, ax=ax2, label='Wind Speed (m/s)')
ax2.set_xlabel('lon')
ax2.set_ylabel('lat')
ax2.set_title('Contourf')
```
